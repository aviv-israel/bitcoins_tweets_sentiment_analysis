```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
file_name = "../datasets/BTC_2019-05-25_2019-06-25__tweets_data_from_file_english_no_duplicates.csv"
```

```python
tweet_data = pd.read_csv(file_name)
```

```python
tweet_data['tweet'] = tweet_data['tweet'].astype('str')
```

```python
tweet_data['date'] = pd.to_datetime(tweet_data['date'], errors='coerce')
```

```python
len(tweet_data)
```

```python
tweet_data = tweet_data.dropna(subset=['date'])
```

```python
len(tweet_data)
```

```python
tweet_data.dtypes
```

## WordClouds

```python
words = ' '.join([text for text in tweet_data['tweet']])
wordCloud = WordCloud(background_color='white', collocations = False, mode="RGB", width=2000 , height=1000).generate(words)
plt.figure(figsize=(20,10))
plt.imshow(wordCloud)
plt.show()
```

```python
plt.figure(figsize=(20,10))
tweet_data.tweet.groupby([tweet_data["date"].dt.year, tweet_data["date"].dt.month, tweet_data["date"].dt.day]).count().plot(kind="bar")
```
